```python

import twint
import pandas as pd
import nest_asyncio

nest_asyncio.apply()

def scrape_and_combine(queries, limit_per_query=6000, since="2024-01-01", until="2026-12-31", filename="tweets_combined.csv"):
    """
    Scrape tweets dari beberapa query lalu gabungkan ke satu CSV.
    queries: list kata kunci/hashtag
    limit_per_query: jumlah maksimal tweet per query
    since/until: rentang waktu
    filename: nama file CSV hasil gabungan
    """
    all_dfs = []

    for q in queries:
        print(f"Scraping query: {q}")
        c = twint.Config()
        c.Search = q
        c.Limit = limit_per_query
        c.Since = since
        c.Until = until
        c.Store_csv = True
        c.Output = f"{q.replace(' ', '_')}.csv"

        twint.run.Search(c)

        # Baca hasil tiap query
        df_q = pd.read_csv(f"{q.replace(' ', '_')}.csv")
        all_dfs.append(df_q)

    # Gabungkan semua hasil
    df_all = pd.concat(all_dfs, ignore_index=True)
    df_all.to_csv(filename, index=False, encoding="utf-8-sig")
    print(f"Scraping selesai. Total {len(df_all)} data tersimpan di {filename}")
    return df_all

# Contoh penggunaan:
queries = [
    "Minato Aqua graduation",
    "湊あくあ 卒業"
]

df = scrape_and_combine(queries, limit_per_query=6000, filename="aqua_graduation_combined.csv")
df.head()
```
