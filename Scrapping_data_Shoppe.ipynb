```python
from google_play_scraper import app, reviews, Sort, reviews_all
import pandas as pd
import numpy as np
```

```python
result, continuation_token = reviews(
    'com.shopee.id',
    lang='id',
    country='id',
    sort=Sort.MOST_RELEVANT,
    count=10000,
)
```

```python
df = pd.DataFrame(result)
df_clean = df[['content','score','at']]

df_clean.to_csv('shopee_review.csv', index=False, encoding='utf-8')
print("Scrapping selesai dan data sudah tersimpan!")
```
