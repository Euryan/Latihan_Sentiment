```python
import snscrape.modules.twitter as sntwitter
import pandas as pd

def scrape_tweets(query, max_tweets=2000, filename="tweets.csv"):
    tweets = []
    for i, tweet in enumerate(sntwitter.TwitterSearchScraper(query).get_items()):
        if i >= max_tweets:
            break
        tweets.append([tweet.date, tweet.user.username, tweet.content])

    # Simpan ke DataFrame
    df = pd.DataFrame(tweets, columns=["Tanggal", "User", "Teks"])

    # Simpan ke CSV
    df.to_csv(filename, index=False, encoding="utf-8-sig")
    print(f"Scraping selesai. {len(df)} data tersimpan di {filename}")
    return df

# Contoh penggunaan:
df = scrape_tweets("Minato Aqua graduation since:2025-01-01 until:2025-12-31",
                   max_tweets=10000,
                   filename="aqua_graduation.csv")

df.head()
```

```python
import snscrape.modules.twitter as sntwitter
import pandas as pd

def scrape_multiple_queries(queries, max_per_query=6000, filename="tweets_combined.csv"):
    """
    Scrape tweets dari beberapa query sekaligus.
    queries: list kata kunci/hashtag
    max_per_query: jumlah maksimal tweet per query
    filename: nama file CSV hasil gabungan
    """
    all_tweets = []

    for q in queries:
        print(f"Scraping query: {q}")
        for i, tweet in enumerate(sntwitter.TwitterSearchScraper(q).get_items()):
            if i >= max_per_query:
                break
            all_tweets.append([tweet.date, tweet.user.username, tweet.content])

    # Simpan ke DataFrame
    df = pd.DataFrame(all_tweets, columns=["Tanggal", "User", "Teks"])
    df.to_csv(filename, index=False, encoding="utf-8-sig")
    print(f"Scraping selesai. Total {len(df)} data tersimpan di {filename}")
    return df

# Contoh penggunaan:
queries = [
    "Minato Aqua graduation since:2025-01-01 until:2025-12-31",
    "湊あくあ 卒業 since:2025-01-01 until:2025-12-31"
]

df = scrape_multiple_queries(queries, max_per_query=6000, filename="aqua_graduation.csv")

df.head()
```
